```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

#from google.colab import files

# CKAN API endpoint
url = "https://data.ca.gov/api/3/action/datastore_search"

# Parameters for the request
params = {
    "resource_id": "e07c5e0b-cace-4b70-9f13-b3e696cd5a99",
    "limit": 100000  # number of rows to fetch (can adjust as needed)
}
```

```python
response = requests.get(url, params=params)
data = response.json()

# Extract the records and convert to DataFrame
records = data["result"]["records"]
df = pd.DataFrame.from_records(records)

# df["Analyte"].unique()

elements_to_include = ['Copper, Total', 'Lead, Total', 'E. coli', 'Enterococcus', 'Iron, Total', 'Manganese, Total', 'Total Suspended Solids, Total', 'Total Dissolved Solids, Total', 'Calcium, Total', 'Nitrate + Nitrite as N, Total']

# Filtering the table to only include the analytes that are in the "elements_to_include" list

df = df[df["Analyte"].isin(elements_to_include)]
```

```python
prism_df = pd.read_csv('/content/PRISM 2024 800 meter data.csv', skiprows = 11)
prism_df.columns = ['Station', 'Longitude', 'Latitude', 'Elevation_ft', 'Date', 'ppt_inches', 'tmin_F', 'tmean_F', 'tmax_F', 'tdmean_F', 'vpdmin_hPa', 'vpdmax_hPa']
merged_df = df.merge(prism_df, on=['Latitude', 'Longitude'], how='left')
# merged_df
filtered_merged_df = merged_df[merged_df['ppt_inches'].notna()]
filtered_merged_df.head(5)
```
